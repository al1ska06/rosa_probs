Problem 1: KMER (k-mer composition)

```python
def lexf(elements, k): #will create a list of every 4-mer
    if k == 1:
        return elements
    else:
        return [
            f"{e}{k_str}"
            for e in elements
            for k_str in lexf(elements, k - 1)
        ]



with open('composition.txt', 'r') as inp:
    seq_string = inp.read().replace('\n', '')[14:]


check = {}
for i in lexf("ACGT", 4):
    occur = 0
    index = seq_string.find(i, 0)
    while index != -1:
        occur +=1
        index = seq_string.find(i, index + 1)
    check[i] = occur

print(*check.values())
```

Problem 2: CORR

```python
reverse_tab = {
    'T' : 'A',
    'A' : 'T', 
    'C' : 'G',
    'G' : 'C'
}

def revComp(str):
    rev = ""
    for t in range(len(str)):
        rev = reverse_tab[str[t]] + rev
    return rev


def hamm(str1, str2):
    amount = 0
    for i in range(len(str1)):
        if str1[i] != str2[i]:
            amount = amount + 1
    return amount
```

```python
with open('errors.txt', 'r') as li:
    li = li.read().replace('\n', '').split('>Rosalind_')
    li = [item[4:] for item in li if item]

total = li + [revComp(item) for item in li]
correct = []
incorrect = []
#for each item in list, check if it is already in correct OR it has another index
for i in range(len(li)):
    if li[i] in correct:
        continue
    try: 
        if total[i+1:].index(li[i]) != -1:
            correct.append(li[i])
    except ValueError:
        incorrect.append(li[i])

#save the reverse complements of the correct ones too
full_correct = correct + [revComp(item) for item in correct]

#for each incorrect one, find the item with a hamming distance of 1
fixed_incorrect = []
for err_seq in incorrect:
    for good_seq in full_correct:
        if hamm(err_seq, good_seq) == 1:
            fixed_incorrect.append((err_seq,good_seq))
            break

for i in range(len(fixed_incorrect)):
    print(fixed_incorrect[i][0] + '->' + fixed_incorrect[i][1])
    

```
