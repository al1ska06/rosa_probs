Problem 1: PMCH (RNA Secondary Structures)

```python
import math 

inp = "ACGCAACGCCUUCAAGUGAUCACUUACGGGAAUUGACGUAGAGUGUUCGUGACCACGCUAGGGUACCGUUCC"
a_count = 0
g_count = 0
for i in inp:
    if i =='A':
        a_count +=1
    if i =='G':
        g_count +=1

#total number of pairings = unique ways to connect a-u * unique ways to connect g-c
print(math.factorial(a_count) * math.factorial(g_count))
```

Problem 3: PROB (intro to random strings)

```python
st = "TAACATCTGCCATCCCAGCCTGCACTTTTTGTACGCTGGGGAACGTGGGCTTTGTGGACCGCTAGTGGAGAACCCTCACCTTGGTTGCCCAGGGCT"
A = open('st.txt', 'r')
A = A.read()
A = A.split(' ')
A = [float(item) for item in A] #A is a list of GC contents

B = []
for gc in A:
    p_g = gc * .5
    p_a = (1-gc) * .5
    prob = 1
    for base in st:
        if base == 'A' or base == 'T':
            prob *= p_a
        if base == 'G' or base =='C':
            prob *= p_g
    prob = math.log(prob, 10)
    B.append(round(prob,4))

print(*B)
```

Problem 4: TRAN (transitions and transversions)

```python
strings = open('trans.txt', 'r')
strings = strings.read()
strings = strings.replace('\n', '')
strings = strings.split('>Rosalind_')
strings = [item[4:] for item in strings if item]


ts = 0
tv = 0
for i in range(len(strings[0])):
    temp1 = strings[0][i]
    temp2 = strings[1][i]
    if temp1 == temp2:
        pass
    elif (temp1 == 'A' or temp1 == 'G') and (temp2 == 'A' or temp2 == 'G'):
        ts +=1
    elif (temp1 == 'C' or temp1 == 'T') and (temp2 == 'C' or temp2== 'T'):
        ts +=1
    else:
        tv +=1

print(ts/tv)
```

Problem 5: TREE (completing a tree)

```python
import re
edge_list = open('tree.txt', 'r')
edge_list = edge_list.read()

#list of all unique nodes in the graph
nums = re.split(r'[\n ]', edge_list)
nums = set(nums)

# couplings of available edges
edge_list = edge_list.split('\n')
edge_list = [item.split(' ') for item in edge_list if item]

print(nums)

def subtree(nodes, edges):
    tree_nodes = nodes.copy()
    for node in nodes: #search for it in the overarching list
        i = 0
        while i < len(edges):
            if(node[0] == edges[i][0]) or (node[0] == edges[i][1]) or (node[1] == edges[i][0]) or (node[1] == edges[i][1]):
                tree_nodes.append(edges[i])
                del edges[i]
            else:
                i +=1
    if len(tree_nodes) == len(nodes): #nothing new was added
        return tree_nodes
    else:
        return(subtree(tree_nodes, edges))

#this part finds number of non-solitary subtrees
n_subs = 0
while edge_list:
    temp = edge_list.pop(0)
    print(subtree([temp], edge_list))
    n_subs +=1

#this part finds remaining values
n_subs += (892 - len(nums))

print(n_subs-1)    
```
