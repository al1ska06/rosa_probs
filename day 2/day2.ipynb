Problem 1: IEV (calculating expected offspring)

```python
f = open("random.txt", 'r')
line = f.read()
num = 0.00
i = 0
while line:
    temp = float(line[:line.index(' ')])
    if i == 3:
        temp *=.75
    if i == 4:
        temp *= .5
    if i == 5:
        temp = 0
    num +=temp
    line = line[line.index(' ')+1 : ]
    i = i+1

print(num * 2)
```

Problem 2: CONS (Consensus and Profile)

```python
#remove rosalinds from fasta format
f = open('mat.txt', 'r')
l = f.read().split(">Rosalind_")
l = [entry.replace('\n', '') for entry in l if entry]
l =[s[4:] for s in l]

print(l)
```

```python
#create empty profile matrix (4xlen(first line))
import numpy as np
```

```python
#while line, temp string is the first line until \n, profile[X, j] 
#gains 1 where X is the letter at index j 
prof = np.zeros((4,len(l[0])))

for seq in l:
    for i in range (len(l[0])):
        if seq[i] == 'A':
            prof[0][i] = prof[0][i] +1
        if seq[i] == 'C':
            prof[1][i] = prof[1][i] +1
        if seq[i] =='G':
            prof[2][i] = prof[2][i] +1
        if seq[i] =='T':
            prof[3][i] = prof[3][i] +1


A = prof[0].astype(int)
C = prof[1].astype(int)
G = prof[2].astype(int)
T = prof[3].astype(int)

final = ""
for i in range(len(A)):
    temp = max(A[i], C[i], G[i], T[i])
    if temp==A[i]:
        final += 'A'
    elif temp == C[i]:
        final +='C'
    elif temp == G[i]:
        final +='G'
    elif temp == T[i]:
        final +='T'

print(final)
print("A:", *A)
print("C:", *C)
print("G:", *G)
print("T:", *T)
```

Problem 3: FIBD (Mortal Fibonacci Rabbits)

```python
def mortal(n, m):
    #start list of rabbits: at first, the first pair is 1 month and the second-max age pair is 0 months
    rabbits = [1] + [0] * (m - 1) 

    for _ in range(1, n):
        # each non-zero rabbit produces 1 new rabbit pair
        new_borns = sum(rabbits[1:])
        
        # drop the last rabbits and add the newly born ones
        rabbits = [new_borns] + rabbits[:-1] 
    return sum(rabbits)
```

Problem 4: GRPH (Overlapping Graphs)

```python
#remove rosalinds from fasta format
f = open('gra.txt', 'r')
l = f.read().split(">")
l = [entry.replace('\n', '') for entry in l if entry]
ids = [s[:13] for s in l]
seqs =[s[13:] for s in l]
```

```python
found = set()
for i in range(len(seqs)): #repeat for all sequences              
    sub = seqs[i][-3:] #find last 3 letters of that sequence
    for k in range(len(seqs)): #check the first 3 letters of every other sequence
        if k!=i and sub == seqs[k][:3]:
            #check if this pair of i,k has been checked. if it has, do not check again
            pair = (i, k)
            if pair not in found:
                found.add(pair)
                print(ids[i], ids[k])
```

Problem 5: LCSM (finding a shared motif)

```python
#extract sequences from fasta
g = open('GCSub.txt', 'r')
seqList = g.read().split(">Rosalind_")
seqList = [entry.replace('\n', '') for entry in seqList if entry]
seqList =[s[4:] for s in seqList]
```

```python
max_string = ""
max_len = 0
#find each substring of the longest item in seqList
dna = max(seqList, key=len)
for i in range(len(dna)):
    k = i
    sub = dna[i]
    while (all(sub in s for s in seqList) ) and k < (len(dna)-1): #all members of seqList
        k = k+1
        sub += dna[k]
    #retruncate sub because last one not in it
    sub = sub[:len(sub)-1]
    #if sub len > max len, make max string that string
    if len(sub) >= max_len:
        max_len = len(sub)
        max_string = sub

print(max_string)
```
