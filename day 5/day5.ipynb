Problem 1: LONG (Genome Assembly as Shortest Superstring)

```python
o = open('seq.txt', 'r')
o = o.read()
o = o.replace('\n', '')
o = o.split('>Rosalind_')
sequences = [item[4:] for item in o if item]

#I'm going to need the largest common substring anyways, so I'll just find that right away

def GCS(old, new): #returns front if the new item has to be placed in front
    countf, countb = 0 , 0
    retf = ""
    retb = ""
    index = 0
    #front sequence:
    while(index < len(new)):
        s1 = old[:index]
        s2 = new[len(new) - index:]
        if (s1==s2):
            countf = index
            retf = new[:len(new)-index]
        index +=1
    #back sequence:
    index = 0
    while(index < len(new)):
        s1 = old[len(old) - index:]
        s2 = new[:index]
        if(s1==s2):
            countb = index
            retb = new[index:]
        index +=1
    if countf > countb:
        return((retf, countf, 'f'))
    elif (countf == 0 and countb == 0):
        return(("none", 0,  'n'))
    else:
        return((retb, countb, 'b'))
    
ret = sequences[0]
del sequences[0]


while sequences:
    most = 0
    to_merge = ""
    typ = ''
    index = -1
    for i in range(len(sequences)):
        result = GCS(ret, sequences[i])
        if result[1] > most:
            most = result[1]
            typ = result[2]
            index = i
            if typ == 'b':
                to_merge = result[0]
            else:
                to_merge = result[0]
    if(index == -1 or most == 0 or to_merge == ""):
        print("we didn't find a suitable merge")
        break
    if typ == 'b':
        ret += to_merge
    else:
        ret = to_merge + ret
    del sequences[index]

print (ret)
```

Problem 3: LEXF (k-mers Lexicography)

```python
def lexf(elements, k):
    if k == 1:
        return elements
    else:
        return [
            f"{e}{k_str}"
            for e in elements
            for k_str in lexf(elements, k - 1)
        ]

for i in lexf("ABCD", 4):
    print(i)
```

Problem 4: LGIS (longest increasing subseq)

```python
#recursive, works on a small scale
def rseq(last, left, current):
    if len(left) == 0:
        return current[:]
    new = [i for i in left if i > last] #possible contenders for next in this substring
    if len(new) == 0:
        return current[:]
    else:
        result = []
        for i in range(len(new)):
            current.append(new[i])
            result.append( rseq(new[i], new[i+1:], current))
            current.pop()
    return max(result, key = len)

def dseq(last, left, current):
    if len(left) == 0:
        return current[:]
    new = [i for i in left if i < last] #possible contenders for next in this substring
    if len(new) == 0:
        return current[:]
    else:
        result = []
        for i in range(len(new)):
            current.append(new[i])
            result.append( dseq(new[i], new[i+1:], current))
            current.pop()
    return max(result, key = len)

wow = [5, 1, 4, 2, 3]
incs = []
decs = []
for i in range(len(wow)):
    incs.append(rseq(wow[i], wow[i+1:], [wow[i]]))
    decs.append(dseq(wow[i], wow[i+1:], [wow[i]]))


print (max(incs, key = len))
print (max(decs, key = len))

```

```python
def lis(arr):
    n = len(arr)
    m = [0] * n
    for x in range(n - 2, -1, -1):
        for y in range(n - 1, x, -1):
            if arr[x] < arr[y] and m[x] <= m[y]:
                m[x] += 1
        max_value = max(m)
        result = []
        for i in range(n):
            if m[i] == max_value:
                result.append(arr[i])
                max_value -= 1
    return result


def lds(arr):
    n = len(arr)
    m = [0] * n
    for x in range(n - 2, -1, -1):
        for y in range(n - 1, x, -1):
            if arr[x] > arr[y] and m[x] <= m[y]:
                m[x] += 1
        max_value = max(m)
        result = []
        for i in range(n):
            if m[i] == max_value:
                result.append(arr[i])
                max_value -= 1
    return result



file = 'rosalind_lgis.txt'
with open(file) as f:
    n = int(f.readline())
    numbers = list(map(int, f.readline().strip().split()))
lis_seq = lis(numbers)
print(' '.join(map(str, lis_seq)))
lds_seq = lds((numbers))
print(' '.join(map(str, lds_seq)))
```

Problem 5: SIGN (signed permutations)

```python
def perm(lis):
    if len(lis)== 1:
        return [lis]
    res = []
    for i in range(len(lis)):
        elem = lis[i]
        rem = lis[:i] + lis[i+1:]
        for p in perm(rem):
            res.append([elem] + p)
    return res

def negate(lis):
    unsigned = perm(lis)
    res = []
    for p in unsigned:
        for signs in product([1,-1], repeat = len(lis)): #product() is an itertools function getting all possible combinations
            signed = []
            for i in range(len(lis)):
                signed.append(p[i] * signs[i])
            res.append(signed)
    return res


final = (negate([1,2,3,4]))
print (len(final))
for i in range(len(final)):
    print(*final[i])
```
