Problem 1: MMCH (Maximum matchings)

```python
import math

#if the maximum base pairs are used, then:
    #for AU pairs, there are a A's and u U's
    #the maximum matching means that there are [max(a,u) P min (a, u)] pairings of a, u

a, u, g, c = 0, 0, 0, 0

for ch in "ACUUAACGUAAAAUGUAAAGUAAAUGUCUCACUGCAGCCGUCGGAUAAUGCAUCAAGUCACGAAUGGCAACCUGACUAAUUUGCUGCCCAUCGGGUCCCC":
    if ch == 'A':
        a+=1
    if ch == 'U':
        u+=1
    if ch =='C':
        c +=1
    if ch =='G':
        g +=1

#AU pairings:
m = max(a, u)
n = min(a,u)
#GC pairings
o = max(g,c)
p = min(g,c)

print((math.perm(m,n)) * (math.perm(o,p)))
```

Problem 4: CAT (Catalan numbers)

```python
#to qualify to be a perfect matching, all pairings must be between an even and odd index
#for all list[0] pairings that satisfy being from list[0] to list[2i+1], check if 
# list[1:2i+1] and list[2i+2:] have such a pairing. 
# when len(list) == 0, return 1

#recursive solution that works slowly
def DNAcat(str):
    if len(str) == 0:
        return 1
    if len(str)%2 == 1:
        print ("at some point, an invalid best string was chosen")
        return -100000000000
    first = str[0]
    second = ''
    match first:
        case 'A':
            second = 'U'
        case 'U':
            second = 'A'
        case 'G':
            second = 'C'
        case 'C':
            second = 'G'
    total = 0
    for i in range(1, len(str), 2): #picks the odd indices
        if str[i] == second:
            total += DNAcat(str[1:i]) * DNAcat(str[i+1:])
    
    return total

print(DNAcat("UUGAAUGCAUUAUCGGCCCGAAUA"))
```

```python
def is_pair(a, b):
    return (a == 'A' and b == 'U') or (a == 'U' and b == 'A') or (a == 'G' and b == 'C') or (a == 'C' and b == 'G')

def DNAcat_dp(s):
    n = len(s)
    dp = [[0] * n for _ in range(n)] #initialize a matrix to save if an initial [i, j] ends up successful
    

    for i in range(n):
        dp[i][i] = 1 
    
    # Consider all substrings of length 2 to n (only even length can pair fully)
    for length in range(2, n+1, 2):
        for i in range(n - length + 1):
            j = i + length - 1
            total = 0
            for k in range(i+1, j+1, 2):
                if is_pair(s[i], s[k]):
                    left = dp[i+1][k-1] if k-1 >= i+1 else 1
                    right = dp[k+1][j] if k+1 <= j else 1
                    total += left * right
            dp[i][j] = total

    return (dp[0][n-1])%1000000 if n > 0 else 1

print(DNAcat_dp("UUGAAUGCAUUAUCGGCCCGAAUA"))
```

Problem 5: LEXV (ordering strings)

```python
inp = "DNA"
length = 3

def addTo(curr, tot, most, depth, res):
    if depth > 0 and depth <= most:
        res.append(curr)
    if depth == most:
        return
    for char in tot:
        addTo(curr + char, tot, most, depth+1, res)
    return res

my_list = addTo("", "CUWLSJRXPEB", 3, 0, [])
for item in my_list:
    print(item)
```
